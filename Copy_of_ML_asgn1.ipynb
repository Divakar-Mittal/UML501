```python
#(g)
p = np.array([[1, 2], [2, 3]])
q = np.array([[4, 5], [6, 7]])

dot_product = np.dot(p, q)
print(dot_product)

###inefficient method :
#combined = np.stack((p.flatten(), q.flatten()), axis = 0)
#cov = np.cov(combined)
#print(cov)

#correct method :
#find p centered about it's mean of the features
p_c = p - p.mean(axis = 0)
#do the same for q
q_c = q - q.mean(axis = 0)
#now dim of p -> n * d1 and q -> n * d2
#now we want the featured to be multiplied so transpose p_c
ans = (p_c.T @ q_c)/(p.shape[0] - 1)
print(ans)
```

```python
def img_to_array(path) :
  #try to read image as gray scale :
  img = cv2.imread(path)

  if img is None :
    print("path not found")
    return

  #check if the image is 2D :
  if len(img.shape) == 2 :
    print("img is gray scale")
    np.savetxt("grayscale_image.txt", img, fmt='%d')
    print("saved as grayscale_image.txt")
  else :
    #read image as color image
    #convert the BGR to RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    print("the image is RGB")
    # Flatten the RGB image to save
    img_rgb_flat = img_rgb.reshape(-1, 3)
    np.savetxt("rgb_image.txt", img_rgb_flat, fmt='%d')
    print("Saved as rgb_image.txt")
```
